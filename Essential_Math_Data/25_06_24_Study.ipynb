### upper is ordinary, below is chain rule. (composite function way)

```python
x,y=symbols('x y')
_y=x**2+1
z=y**3-2
dy_dx=diff(_y,x)
dz_dy=diff(z,y)
dz_dx=(dy_dx*dz_dy).subs(_y,x**2+1).subs(y,x**2+1)
print(dz_dx)
```

```python
def approximate_integral(a,b,n,f): # integral approximate
    dalta_x=(b-a)/n
    total_sum=0
    for i in range(n):
        total_sum+=dalta_x*f(a+dalta_x*(i+1/2))
    return total_sum
def formula(x):
    return x**2+1
area=approximate_integral(0,1,10000,formula)
print(area)
```

```python
x,i,n=symbols('x i n')

f=x**2+1
lower,upper=0,7
delta_x=((upper-lower)/n)
x_i=(lower+delta_x*i)
fx_i=f.subs(x,x_i)

n_rectangle=Sum(delta_x*fx_i,(i,1,n)).doit()
ara=limit(n_rectangle,n,oo)

print(ara)
```

### nomal distribution
- bell shape
- both tails become smaller
- Bigger in around of mean.

```python
from scipy.stats import norm #mean place middle.

mean=64.43
std_dev=2.99

x=norm.cdf(64.43,mean,std_dev)
print(x)
```

```python
from scipy.stats import norm #opposite of cdf.

x=norm.ppf(0.40,loc=64.43,scale=2.99)
print(x)
```

```python
from scipy.stats import norm
def criticalZ(p):
    data=norm(loc=0,scale=1)
    leftArea=(1-p)/2
    rightArea=1-leftArea
    return data.ppf(leftArea),data.ppf(rightArea)
print(criticalZ(0.95))
```
