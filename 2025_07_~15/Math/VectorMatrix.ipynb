````python
import numpy as np
import matplotlib.pyplot as plt

def ShowVectors(*Vectors):
    # 벡터 정의
    # 원점
    origin = np.array([0, 0])

    plt.figure(figsize=(6, 6))
    colors=['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
    for i,ve in enumerate(Vectors):
        plt.quiver(
            origin[0], origin[1],       # 시작점 x, y (여기선 모두 원점)
            ve[0], ve[1],               # v1의 x, y 성분
            angles='xy', scale_units='xy', scale=1, color=colors[i], label='v'+f'{i}'
        )
    # 좌표 범위와 눈금 설정
    plt.xlim(-4, 10)
    plt.ylim(-4, 10)
    plt.xticks(range(-4, 10, 1))    # 격자설정
    plt.yticks(range(-4, 10, 1))    # 
    plt.grid(True)
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    plt.gca().set_aspect('equal')
    plt.legend()
    plt.show()
````

## ✅ 한 줄 요약

> **BLAS**는 **기초 선형대수 연산**을 위한 표준 라이브러리이고,
> `NumPy`, `PyTorch`, `TensorFlow` 같은 과학 컴퓨팅 라이브러리는 **이걸 내부적으로 사용해 연산을 빠르게 처리**합니다.
> 이를 흔히 “BLAS 가속”이라고 합니다.

---

## 📚 BLAS란?

| 항목     | 설명                                                      |
| ------ | ------------------------------------------------------- |
| **이름** | **B**asic **L**inear **A**lgebra **S**ubprograms        |
| **역할** | 선형대수 연산(행렬곱, 벡터곱 등)을 빠르게 수행                             |
| **내용** | `dot`, `matmul`, `norm`, `gemm`, `axpy` 같은 저수준 연산 함수 모음 |
| **형태** | 인터페이스(표준) → 여러 구현체(LAPACK, OpenBLAS, Intel MKL 등)가 있음   |

---

## 🔧 실제로는 어떻게 쓰이나?

### 예: `NumPy`에서의 행렬 곱

```python
import numpy as np

A = np.random.rand(1000, 1000)
B = np.random.rand(1000, 1000)

C = A @ B
```

* 위 연산은 단순한 `for` 루프가 아니라,
* \*\*BLAS 라이브러리 (예: Intel MKL, OpenBLAS)\*\*에서 제공하는 최적화된 C 코드로 실행됨
  → 그래서 **몇 배 \~ 수십 배 빠름**

---

## 🧠 어떤 연산이 BLAS 가속 대상인가?

| 수준      | 연산 종류    | 예시                                        |
| ------- | -------- | ----------------------------------------- |
| Level 1 | 벡터 연산    | 내적, 스칼라곱, 노름 (`np.dot`, `np.linalg.norm`) |
| Level 2 | 행렬-벡터 연산 | `A @ x` (행렬 × 벡터)                         |
| Level 3 | 행렬-행렬 연산 | `A @ B` (행렬 × 행렬)                         |

→ **Level 3 연산이 가장 많은 최적화 효과**를 누림 (CPU, 캐시 병렬화 등)

---

| 효과   | 벡터, 행렬 곱, 선형방정식 풀이 등 연산 속도 향상 (수십 배 빠를 수도 있음) |
